```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data_final/renewable-energy-generation.csv')


filtered_df = df.rename(columns={'Entity': 'Country'})
filtered_df = filtered_df.sort_values(by='Country')
filtered_df
```

```python
print(f'Data shape: {filtered_df.shape[0]} rows, {filtered_df.shape[1]} columns')
```

```python
# Get the distinct countries
distinct_countries_count = filtered_df['Country'].nunique()
distinct_countries = filtered_df['Country'].unique()

# Display the result
print(f'There are {distinct_countries_count} countries in the dataset')
print(f'Countries: {distinct_countries}')

```

```python
# Display null values
sns.heatmap(filtered_df.isnull(), yticklabels = False, cbar = False, cmap = 'rainbow')

# Count null values
filtered_df.isnull().sum()
```

```python
# Drop code column

filtered_df = filtered_df.drop(labels='Code',axis=1)
filtered_df
```

```python

filtered_df.describe()
```

```python
# Save file for using for forecast
filtered_df.to_csv('../data_for_forecast/renewable_energy_generation-clean.csv')
filtered_df
```

```python
# Countries with all 4 columns with zero values
filtered_df_countries_with_zero_values = filtered_df[(filtered_df[['Other renewables (including geothermal and biomass) electricity generation - TWh', 'Solar generation - TWh', 'Wind generation - TWh', 'Hydro generation - TWh']] == 0).all(axis=1)]
countries_zero_values = filtered_df_countries_with_zero_values['Country'].unique()
print(f'The following countries do not generate any renewable energy or the data is missing:\n{countries_zero_values}')
```

```python
# TREND ANALYSIS OVER TIME - RENEWABLE ENERGY GENERATION OVER THE YEARS
# Hydro generation - TWh	

yearly_fec = filtered_df.groupby('Year')['Hydro generation - TWh'].mean()
plt.plot(yearly_fec.index, yearly_fec.values, marker='o')
plt.xticks([2019,2020,2021,2022,2023])
plt.title('Hydro generation in Europe (2019 - 2023)')
plt.xlabel('Year')
plt.ylabel('Hydro generation (TWh)')
plt.show()
```

```python
# TOP COUNTRIES WITH THE MOST RENEWABLE ENERGY GENERATION IN 2023
# Hydro generation - TWh	

data_2023 = filtered_df[filtered_df['Year'] == 2023]

top_countries = data_2023[['Country', 'Hydro generation - TWh']].sort_values(by='Hydro generation - TWh', ascending=False)
print(top_countries.head(10))

plt.barh(top_countries['Country'][:10],top_countries['Hydro generation - TWh'][:10], color = 'green')
plt.title('Top 10 Countries with the most Hydro generation in 2023')
plt.xlabel('Hydro generation (TWh)')
plt.ylabel('Country')
plt.show()

```

```python
# Renewable energy generation - higher % of -> better sustainability 
# Hydro generation - TWh
avg_renewable_hydro = filtered_df.groupby('Country')['Hydro generation - TWh'].mean().reset_index()

max_value = avg_renewable_hydro['Hydro generation - TWh'].max()
min_value = avg_renewable_hydro['Hydro generation - TWh'].min()

avg_renewable_hydro['renewable_hydro_generation_subindex'] = (avg_renewable_hydro['Hydro generation - TWh'] - min_value)/(max_value - min_value)

avg_renewable_hydro.sort_values(by='renewable_hydro_generation_subindex', ascending=False)
```

```python
avg_renewable_hydro.to_csv('../data_subindex/subindex_renewable_hydro_energy_generation_2019-2023.csv')
```

```python
avg_renewable_hydro = avg_renewable_hydro.sort_values(by='renewable_hydro_generation_subindex')

plt.figure(figsize=(10,15))
plt.barh(avg_renewable_hydro['Country'],avg_renewable_hydro['renewable_hydro_generation_subindex'], color='green')
plt.xlabel('0 = worst, 1 = best')
plt.title('Renewable Hydro Energy Generation Subindex')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Renewable energy generation - higher % of -> better sustainability 
# Solar generation - TWh
avg_renewable_solar = filtered_df.groupby('Country')['Solar generation - TWh'].mean().reset_index()

max_value = avg_renewable_solar['Solar generation - TWh'].max()
min_value = avg_renewable_solar['Solar generation - TWh'].min()

avg_renewable_solar['renewable_solar_generation_subindex'] = (avg_renewable_solar['Solar generation - TWh'] - min_value)/(max_value - min_value)
avg_renewable_solar['renewable_solar_generation_subindex']  = avg_renewable_solar['renewable_solar_generation_subindex'].round(5)

avg_renewable_solar.sort_values(by='renewable_solar_generation_subindex', ascending=False)
```
